# Aggregation and Grouping
> **groupby** tells how to group the data.  
> Operations to be applied to the groups: **mean**, **min**, etc.  

````python
import numpy as np
import pandas as pd

rng = np.random.RandomState(42)
x = pd.Series(rng.rand(5))
print(x)
print('--------------')
print(x.sum())
print('--------------')
print(x.mean())
````

````python
df = pd.DataFrame({'A': rng.rand(5),
                   'B': rng.rand(5)})
print(df)
print('--------------')
print(df.mean()) # by variable/ column
print('--------------')
print(df.mean(axis=0))
````

````python
print(df.mean(axis=1))
print('--------------')
````

````python
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data': range(6)}, columns=['key', 'data'])
df
````

````python
df.groupby('key').sum()
````

> 2. Group by the **key** variable.  
> 3. For each **key**, obtain **min**.  
> 4. For each **key**, obtain **max**.  
> 5. For each **key**, obtain **mean**.  

````python
import numpy as np
import pandas as pd

rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'data1': range(6),
                   'data2': rng.randint(0, 10, 6)},
                  columns = ['key', 'data1', 'data2'])
df.head()
````

````python
df.groupby('key').min()
````

````python
df.groupby('key').max()
````

````python
df.groupby('key').mean()
````

## Aggregation

````python
rng = np.random.RandomState(0)
df = pd.DataFrame({'key': ['A', 'B', 'C', 'A', 'B', 'C'],
                'data1': range(6),
                'data2': rng.randint(0, 10, 6)},
               columns = ['key', 'data1', 'data2'])
df
````

````python
df.groupby('key').aggregate([min, np.mean, max])
````

## Transformation

````python
df.groupby('key').transform(lambda x: x - x.mean())
````

````python
decade = 10 * (planets['year'] // 10)
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method', decade])['number'].sum().unstack()
````
